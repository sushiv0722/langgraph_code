```python
from langgraph.graph import StateGraph ,START , END 
from typing import TypedDict
```

```python
##creating state 

class bmi_state(TypedDict):

    weight_kg:float
    height_m:float
    bmi:float
    
```

```python
## defining then graph 

graph=StateGraph(bmi_state)

```

```python
def calculate_bmi(state:bmi_state)->bmi_state:
    weight=state['weight_kg']
    height=state['height_m']

    bmi=weight/(height**2)

    state['bmi']=round(bmi,2)

    return state
```

```python
## creating node in graph 

graph.add_node("calculate_bmi_node",calculate_bmi)
```

```python
## creating then edge 

graph.add_edge(START,"calculate_bmi_node")
graph.add_edge("calculate_bmi_node",END)
```

```python
workflow=graph.compile()
```

```python
initial_state={'weight_kg':80,'height_m':1.73}

output_state=workflow.invoke(initial_state)

print(output_state)
```
